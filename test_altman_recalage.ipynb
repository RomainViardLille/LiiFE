```python
import SimpleITK as sitk
import nibabel as nib
import numpy as np
from scipy.ndimage import zoom
import matplotlib.pyplot as plt
import os
```

```python
def bland_altman_plot(image1, image2):
    # Calculate the mean and the difference between the two images
    mean = (image1.get_fdata() + image2.get_fdata()) / 2
    diff = image1.get_fdata() - image2.get_fdata()

    # Create the Bland-Altman plot
    plt.figure(figsize=(8, 6))
    #plt.scatter(mean, diff, s=20, c='blue', alpha=0.5)
    #plt.axhline(y=np.mean(diff), color='red', linestyle='--')
    plt.scatter(mean, diff, alpha=0.5)
    plt.axhline(np.mean(diff), color='red', linestyle='--', label='Mean Difference')
    plt.axhline(1.96 * np.std(diff), color='gray', linestyle='--', label='Upper Limit of Agreement')
    plt.axhline(-1.96 * np.std(diff), color='gray', linestyle='--', label='Lower Limit of Agreement')
    plt.xlabel('Mean of Images 1 and 2')
    plt.ylabel('Difference (Image 1 - Image 2)')
    plt.title('Bland-Altman Plot')
    plt.grid(True)

    # Show the plot
    plt.show()

    
    
    
    """""# Create the Bland-Altman plot
    plt.figure(figsize=(8, 6))
    plt.scatter(mean, diff, alpha=0.5)
    plt.axhline(np.mean(diff), color='red', linestyle='--', label='Mean Difference')
    plt.axhline(1.96 * np.std(diff), color='gray', linestyle='--', label='Upper Limit of Agreement')
    plt.axhline(-1.96 * np.std(diff), color='gray', linestyle='--', label='Lower Limit of Agreement')

    # Customize the plot
    plt.xlabel('Mean of Measurements')
    plt.ylabel('Difference between Measurements')
    plt.title('Bland-Altman Plot')
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()"""""
```

```python
def bland_altman_plot2(image1, image2):
    # Get image data as numpy arrays
    data1 = image1.get_fdata()
    data2 = image2.get_fdata()

    # Calculate the mean and the difference between the two images
    mean = (data1 + data2) / 2
    diff = data1 - data2

    # Create the Bland-Altman plot
    plt.figure(figsize=(8, 6))
    plt.scatter(mean, diff, alpha=0.5)
    plt.axhline(np.mean(diff), color='red', linestyle='--', label='Mean Difference')
    plt.axhline(1.96 * np.std(diff), color='gray', linestyle='--', label='Upper Limit of Agreement')
    plt.axhline(-1.96 * np.std(diff), color='gray', linestyle='--', label='Lower Limit of Agreement')

    # Customize the plot
    plt.xlabel('Mean of Measurements')
    plt.ylabel('Difference between Measurements')
    plt.title('Bland-Altman Plot')
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()
```

```python
""""
def rigid_registration(image_fixed_path, image_moving_path):
    # Load NIfTI images using nibabel
    image_fixed = nib.load(image_fixed_path)
    image_moving = nib.load(image_moving_path)

    # Get the data arrays from the images
    data_fixed = image_fixed.get_fdata()
    data_moving = image_moving.get_fdata()

    # Convert the data arrays to SimpleITK images
    sitk_fixed = sitk.GetImageFromArray(data_fixed)
    sitk_moving = sitk.GetImageFromArray(data_moving)

    # Convert images to float32 data type
    sitk_fixed = sitk.Cast(sitk_fixed, sitk.sitkFloat32)
    sitk_moving = sitk.Cast(sitk_moving, sitk.sitkFloat32)

    # Create a registration object
    registration = sitk.ImageRegistrationMethod()

    # Initialize the registration with a rigid transformation
    initial_transform = sitk.CenteredTransformInitializer(sitk_fixed, sitk_moving, sitk.Euler3DTransform())
    registration.SetInitialTransform(initial_transform)

    # Choose the similarity metric (e.g., mean squares)
    registration.SetMetricAsMeanSquares()

    # Choose the optimizer (e.g., gradient descent)
    registration.SetOptimizerAsRegularStepGradientDescent(learningRate=1.0, minStep=0.01, numberOfIterations=200)

    # Set the interpolator (e.g., linear)
    registration.SetInterpolator(sitk.sitkLinear)

    # Perform the registration to align the moving image with the fixed image
    final_transform = registration.Execute(sitk_fixed, sitk_moving)

    # Resample the moving image to the space of the fixed image using the final transformation
    resampler = sitk.ResampleImageFilter()
    resampler.SetReferenceImage(sitk_fixed)
    resampler.SetTransform(final_transform)

    # Get the registered image as a SimpleITK image
    registered_image_sitk = resampler.Execute(sitk_moving)

    # Convert the registered image to a NIfTI object
    registered_image_nifti = nib.Nifti1Image(sitk.GetArrayFromImage(registered_image_sitk), image_fixed.affine)

    # Save the registered image to a temporary file using nibabel
    output_path = "registered_image.nii.gz"
    nib.save(registered_image_nifti, output_path)

    # Load the registered image using nibabel
    registered_image = nib.load(output_path)

    # Optionally, you can remove the temporary file
    os.remove(output_path)

    return registered_image
```

```python
def recalage(image_fixe, image_mobile):
    image_fixe = nib.load(image_fixe)
    image_mobile = nib.load(image_mobile)

    donnees_fixes = image_fixe.get_fdata()
    donnees_mobiles = image_mobile.get_fdata()

    sitk_fixe = sitk.GetImageFromArray(donnees_fixes)
    sitk_mobile = sitk.GetImageFromArray(donnees_mobiles)

    sitk_fixe = sitk.Cast(sitk_fixe, sitk.sitkFloat32)
    sitk_mobile = sitk.Cast(sitk_mobile, sitk.sitkFloat32)

    registration = sitk.ImageRegistrationMethod()

    initial_transform = sitk.CenteredTransformInitializer(sitk_fixe, sitk_mobile, sitk.Euler3DTransform())
    registration.SetInitialTransform(initial_transform)

    registration.SetMetricAsMeanSquares()

    registration.SetOptimizerAsRegularStepGradientDescent(learningRate=1.0, minStep=0.01, numberOfIterations=200)

    registration.SetInterpolator(sitk.sitkLinear)

    final_transform = registration.Execute(sitk_fixe, sitk_mobile)

    resampler = sitk.ResampleImageFilter()
    resampler.SetReferenceImage(sitk_fixe)
    resampler.SetTransform(final_transform)

    #image_recalee_nifti = nib.Nifti1Image(sitk.GetArrayFromImage(resampler.Execute(sitk_mobile)), image_fixe.affine)
    image = sitk.GetArrayFromImage(resampler.Execute(sitk_mobile)), image_fixe.affine

    #return image_recalee_nifti
    return image
```
