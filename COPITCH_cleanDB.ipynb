```python
import numpy as np
import pandas as pd
import os,sys
import matplotlib.pyplot as plt
import nibabel as nib
import glob

%matplotlib inline
```

```python
STUDY_PATH=os.path.join('/NAS','dumbo','protocoles','copitch')
```

```python
# Define the pattern to search for
pattern = os.path.join(STUDY_PATH, 'data', 'proceed', 'sub*', '*', '*FLAIR3D*.nii.gz')

# Retrieve the list of files
image_files = glob.glob(pattern)

# Loop through each file in image_files
for file in image_files:
    # Get the directory of the current file
    nifti_img = nib.load(file)
    # Extract the two last directories from the file path
    session=os.path.join(*file.split(os.sep)[-2:-1])
    sujet=os.path.join(*file.split(os.sep)[-3:-2])

    ndarray_same_size = np.zeros(nifti_img.shape)

    directory = os.path.dirname(file)
    print(directory)


    # Retrieve the list of new files matching the pattern
    matched_files = glob.glob(os.path.join(directory, '*ICH_PHO*.nii.gz'))
    matched_files.reverse() #afin que les ctrl soient avant les MAJUSCULES
    print(matched_files)
    # If there are matched files, add them to the new_files list
    if matched_files:
        nifti_img = nib.load(matched_files[0])
        data = nifti_img.get_fdata()
        # Check if the shape of ndarray_same_size and data are identical
        if ndarray_same_size.shape == data.shape:
            # Create a mask for the values 1, 2, and 6
            mask = np.isin(data, [1, 2, 6])
            # Set the corresponding values in ndarray_same_size
            ndarray_same_size[mask] = data[mask]
        else:
            print(f"Shape mismatch: ndarray_same_size {ndarray_same_size.shape}, data {data.shape}")

    matched_files = glob.glob(os.path.join(directory, '*hemolys*.nii.gz'))
    matched_files.reverse() #afin que les ctrl soient avant les MAJUSCULES
    if matched_files:
        nifti_img = nib.load(matched_files[0])
        data = nifti_img.get_fdata()
       # Check if the shape of ndarray_same_size and data are identical
        if ndarray_same_size.shape == data.shape:
            # Create a mask for the values 1, 2, and 6
            mask = np.isin(data, [1, 2, 6])
            # Set the corresponding values in ndarray_same_size
            ndarray_same_size[mask] = data[mask]
        else:
            print(f"Shape mismatch: ndarray_same_size {ndarray_same_size.shape}, data {data.shape}")

    # Modify the values 6 to 3 in ndarray_same_size
    ndarray_same_size[ndarray_same_size == 6] = 3
    num_values=np.unique(ndarray_same_size)
    if len(num_values) > 1:
        new_file = os.path.join(STUDY_PATH,'CoNeMOS','CoNeMOS','labels_dir',f'ICH_PHO_HEMO_{sujet}_{session}.nii.gz')
        new_nifti_img = nib.Nifti1Image(ndarray_same_size, nifti_img.affine, nifti_img.header)
        nib.save(new_nifti_img, new_file)

        new_file = os.path.join(STUDY_PATH,'CoNeMOS','CoNeMOS','images_dir',f'FLAIR3D_{sujet}_{session}.nii.gz')
        os.system(f'cp {file} {new_file}')

        print(f"{sujet} {session} {len(num_values)}:{num_values}")
```
