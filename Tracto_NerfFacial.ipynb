```python
def display_mri(data, title=None, cmap="gray", vmin=-1, vmax=0.8): 
    vmin=np.min(data)
    vmax=np.max(data)
    fig,axs = plt.subplots(nrows=1, ncols=5, figsize=(20,8))
    for ax in axs:
        ax.set_xticks(())
        ax.set_yticks(())
    if title:
        fig.suptitle(title)
    axs[0].imshow(data[:,:,5], cmap=cmap, vmin=vmin, vmax=vmax)
    axs[1].imshow(data[:,:,9], cmap=cmap, vmin=vmin, vmax=vmax)
    axs[2].imshow(data[:,:,13], cmap=cmap, vmin=vmin, vmax=vmax)
    axs[3].imshow(data[:,:,17], cmap=cmap, vmin=vmin, vmax=vmax)
    axs[4].imshow(data[:,:,21], cmap=cmap, vmin=vmin, vmax=vmax)    
    fig.tight_layout()
    plt.show()
    plt.savefig(join(PATH_DIR,'sliceofColorCodedVect.jpeg'))
```
