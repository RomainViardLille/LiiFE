```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms

# Define your model class
class MyModel(nn.Module):
    # ...

# Load and preprocess data
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
train_dataset = datasets.MNIST('./data', train=True, transform=transform, download=True)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)

# Initialize the model, loss function, and optimizer
model = MyModel()
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training loop
for epoch in range(10):
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs.view(inputs.size(0), -1))
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

# Specify the full path to save the model
save_path = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Diffusion_model/results_training/savedmodelxxx.pt"

# Save the trained model to the specified path
torch.save(model.state_dict(), save_path)
```

```python
import os
import glob
import nibabel as nib
import numpy as np
from skimage.transform import resize

# Define the input and output folders
input_folder = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Diffusion_model/data/test/174_test"
output_folder = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Diffusion_model/data/resized_174_test"

# Define the target shape
target_shape = (256, 480)

# Recursively traverse the input folder and resize each NIfTI image
for root, dirs, files in os.walk(input_folder):
    for file in files:
        if file.endswith(".nii.gz"):
            # Get the relative path of the file
            relative_path = os.path.relpath(root, input_folder)
            # Construct the output directory path
            output_dir = os.path.join(output_folder, relative_path)
            # Create the output directory if it doesn't exist
            os.makedirs(output_dir, exist_ok=True)
            # Construct the input and output file paths
            input_path = os.path.join(root, file)
            output_path = os.path.join(output_dir, file)
            # Load the NIfTI image
            nifti_img = nib.load(input_path)
            # Get the image data
            img_data = nifti_img.get_fdata()
            # Resize the image
            resized_img_data = resize(img_data, target_shape, anti_aliasing=True)
            # Create a new NIfTI image with the resized data
            resized_nifti_img = nib.Nifti1Image(resized_img_data, nifti_img.affine, nifti_img.header)
            # Save the resized NIfTI image
            nib.save(resized_nifti_img, output_path)

print("Image resizing completed.")
```

```python
import os
import nibabel as nib
import numpy as np

# Define the input and output folders
input_folder = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Diffusion_model/data/final_test_col"
output_folder = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Diffusion_model/data/final_test_col_chann2"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Function to process an individual NIfTI image
def process_nifti(input_path, output_path):
    nifti_image = nib.load(input_path)
    image_data = nifti_image.get_fdata()

    if image_data.shape[-1] == 3:
        modified_image_data = image_data[:, :, :, :2]

        modified_nifti_image = nib.Nifti1Image(modified_image_data, nifti_image.affine)
        nib.save(modified_nifti_image, output_path)

# Walk through the input folder and its subfolders
for root, dirs, files in os.walk(input_folder):
    for file in files:
        if file.endswith(".nii.gz"):
            input_path = os.path.join(root, file)
            relative_path = os.path.relpath(input_path, input_folder)
            output_path = os.path.join(output_folder, relative_path)
            
            # Create the output folder structure if it doesn't exist
            output_subfolder = os.path.dirname(output_path)
            if not os.path.exists(output_subfolder):
                os.makedirs(output_subfolder)
            
            # Process the NIfTI image and save it to the output folder
            process_nifti(input_path, output_path)

print("Processing complete.")
```

```python
import os
import nibabel as nib
import numpy as np

# Define the input and output folders
input_folder = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Diffusion_model/data/final_test_col"
output_folder = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Diffusion_model/data/final_test_col_chann2"

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Function to process an individual NIfTI image and save it to the output directory
def process_and_save_nifti(input_path, output_path):
    nifti_image = nib.load(input_path)
    image_data = nifti_image.get_fdata()

    if image_data.shape[-1] == 3:
        modified_image_data = image_data[:, :, :, :2]

        # Create the output directory structure
        output_dir = os.path.dirname(output_path)
        os.makedirs(output_dir, exist_ok=True)

        modified_nifti_image = nib.Nifti1Image(modified_image_data, nifti_image.affine)
        nib.save(modified_nifti_image, output_path)

# Process NIfTI images in the input directory and save in the output directory
for root, dirs, files in os.walk(input_folder):
    for file in files:
        if file.endswith(".nii.gz"):
            input_path = os.path.join(root, file)
            relative_path = os.path.relpath(input_path, input_folder)
            output_path = os.path.join(output_folder, relative_path)

            # Process the NIfTI image and save it to the output folder
            process_and_save_nifti(input_path, output_path)

print("Processing complete.")

```

```python

import nibabel as nib
import numpy as np
from scipy.ndimage import zoom

# Define the input file paths
t1_input_path = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Before_prettt_T1Gd/CHU/Pat9_T1Gado_CHU.nii.gz"
t2_input_path = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/haca3/Pat9_T1Gado_COL_registred.nii.gz"

# Define the output directory
output_dir = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/haca3/resized"

# Define the target shape
target_shape = (192, 224, 192)

# Load and resize T1 image
t1_img = nib.load(t1_input_path)
t1_data = t1_img.get_fdata()
t1_resized = zoom(t1_data, (target_shape[0] / t1_data.shape[0], target_shape[1] / t1_data.shape[1], target_shape[2] / t1_data.shape[2]), order=3)

# Save the resized T1 image
t1_resized_nifti = nib.Nifti1Image(t1_resized, affine=t1_img.affine)
t1_output_path = f"{output_dir}/Pat8_CHU_T1Gado_resized.nii.gz"
nib.save(t1_resized_nifti, t1_output_path)

# Load and resize T2 image (similar to T1)
t2_img = nib.load(t2_input_path)
t2_data = t2_img.get_fdata()
t2_resized = zoom(t2_data, (target_shape[0] / t2_data.shape[0], target_shape[1] / t2_data.shape[1], target_shape[2] / t2_data.shape[2]), order=3)

# Save the resized T2 image
t2_resized_nifti = nib.Nifti1Image(t2_resized, affine=t2_img.affine)
t2_output_path = f"{output_dir}/Pat8_COL_T1Gado_resized.nii.gz"
nib.save(t2_resized_nifti, t2_output_path)

print("Resized images saved to:", output_dir)
```
