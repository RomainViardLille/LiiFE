```python
import os
import numpy as np
import matplotlib.pyplot as plt
import nibabel as nib
from skimage.util import montage 
from skimage.transform import rotate
%matplotlib inline
#%matplotlib notebook
```

```python
CHU_file="/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/haca3/haca3_example2/test_chu.nii.gz"
chu=np.asanyarray(nib.load(CHU_file).dataobj)

COL_file="/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/haca3/haca3_example2/test_col.nii.gz"
col=np.asanyarray(nib.load(COL_file).dataobj)

H_file="/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/haca3/haca3_example2/test_result_harmonized_fusion.nii.gz"
harm=np.asanyarray(nib.load(H_file).dataobj)
```

```python
fig=plt.figure()

tmp=np.swapaxes(chu[:,:,50:130:20],0,2) 
tmp2=montage(tmp,grid_shape=(1,tmp.shape[0]))
ax1 = fig.add_subplot(3,1,1)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.imshow(tmp2,cmap ='gray',vmin=np.quantile(tmp2,0.01),vmax=np.quantile(tmp2,0.99))
ax1.set_xlabel("CHU - SIGNA GE 1.5T")

tmp=np.swapaxes(col[:,:,50:130:20],0,2) 
tmp2=montage(tmp,grid_shape=(1,tmp.shape[0]))
ax1 = fig.add_subplot(3,1,2)
ax1.set_xticks([])
ax1.set_yticks([])
#ax1.imshow(tmp2,cmap ='gray')
ax1.imshow(tmp2,cmap ='gray',vmin=np.quantile(tmp2,0.01),vmax=np.quantile(tmp2,0.99))
ax1.set_xlabel("COL - SOLA Siemens 1.5T")

tmp=np.swapaxes(harm[:,:,50:130:20],0,2) 
tmp2=montage(tmp,grid_shape=(1,tmp.shape[0]))
ax1 = fig.add_subplot(3,1,3)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.imshow(tmp2,cmap ='gray',vmin=np.quantile(tmp2,0.01),vmax=np.quantile(tmp2,0.99))
ax1.set_xlabel("Images harmonisées CHU -> COL")

#plt.show()
plt.savefig("/NAS/dumbo/protocoles/HAMSI/screenshots/haca3_pt8.png")
#plt.savefig("haca3_pt8.png")
```
