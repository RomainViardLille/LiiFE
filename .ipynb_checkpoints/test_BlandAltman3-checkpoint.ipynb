```python
import nibabel as nib
import numpy as np
from scipy.ndimage import zoom
import matplotlib.pyplot as plt
```

```python
""""
def resample_and_crop_images(image1, image2):
    # Get image data as numpy arrays
    data1 = image1.get_fdata()
    data2 = image2.get_fdata()

    # Ensure both images have the same shape
    if data1.shape != data2.shape:
        # Calculate the target shape as the minimum shape along each dimension
        target_shape = tuple(min(d1, d2) for d1, d2 in zip(data1.shape, data2.shape))

        # Calculate the scaling factors for resampling
        scale_factors = [t / c for t, c in zip(target_shape, data1.shape)]

        # Perform resampling for both images
        resampled_data1 = zoom(data1, scale_factors)
        resampled_data2 = zoom(data2, scale_factors)

        # Create new NIfTI images with the resampled data
        resampled_image1 = nib.Nifti1Image(resampled_data1, affine=image1.affine)
        resampled_image2 = nib.Nifti1Image(resampled_data2, affine=image2.affine)

        # Calculate cropping indices to keep the center of the images
        crop_indices1 = tuple(slice((d1 - t) // 2, (d1 + t) // 2) for d1, t in zip(resampled_data1.shape, target_shape))
        crop_indices2 = tuple(slice((d2 - t) // 2, (d2 + t) // 2) for d2, t in zip(resampled_data2.shape, target_shape))

        # Crop both images to the same shape
        cropped_data1 = resampled_data1[crop_indices1]
        cropped_data2 = resampled_data2[crop_indices2]

        # Create new NIfTI images with the cropped data
        cropped_image1 = nib.Nifti1Image(cropped_data1, affine=image1.affine)
        cropped_image2 = nib.Nifti1Image(cropped_data2, affine=image2.affine)

        return cropped_image1, cropped_image2
    else:
        return image1, image2
```

```python
def crop_to_smaller_shape(image1, image2):
    # Load NIfTI images
    data1 = image1.get_fdata()
    data2 = image2.get_fdata()

    # Get the shape of both images
    shape1 = data1.shape
    shape2 = data2.shape

    # Find the minimum shape along each dimension
    target_shape = tuple(min(s1, s2) for s1, s2 in zip(shape1, shape2))

    # Calculate cropping indices to match the smaller shape
    crop_indices1 = tuple(slice((s1 - t) // 2, (s1 + t) // 2) for s1, t in zip(shape1, target_shape))
    crop_indices2 = tuple(slice((s2 - t) // 2, (s2 + t) // 2) for s2, t in zip(shape2, target_shape))

    # Crop both images
    cropped_image1 = nib.Nifti1Image(data1[crop_indices1], affine=image1.affine)
    cropped_image2 = nib.Nifti1Image(data2[crop_indices2], affine=image2.affine)

    return cropped_image1, cropped_image2
```

```python
def bland_altman_plot(image1, image2):
    # Get image data as numpy arrays
    data1 = image1.get_fdata()
    data2 = image2.get_fdata()

    # Calculate the mean and the difference between the two images
    mean = (data1 + data2) / 2
    diff = data1 - data2

    # Create the Bland-Altman plot
    plt.figure(figsize=(8, 6))
    plt.scatter(mean, diff, alpha=0.5)
    plt.axhline(np.mean(diff), color='red', linestyle='--', label='Mean Difference')
    plt.axhline(1.96 * np.std(diff), color='gray', linestyle='--', label='Upper Limit of Agreement')
    plt.axhline(-1.96 * np.std(diff), color='gray', linestyle='--', label='Lower Limit of Agreement')

    # Customize the plot
    plt.xlabel('Mean of Measurements')
    plt.ylabel('Difference between Measurements')
    plt.title('Bland-Altman Plot')
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()
```
