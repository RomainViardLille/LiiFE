```python
def getNbAvailable(idvb,pwd):
    ladate=datetime.datetime.today().strftime('%Y-%m-%d')
    options = Options()
    options.headless = True
    driver = Firefox(options=options)
    driver.implicitly_wait(15)
    #WebDriverWait(driver, timeout=5).until(document_initialised)
    driver.get("http://volbrain.upv.es/index.php")

    elem = driver.find_element_by_name("email")
    elem.send_keys(idvb)

    elem = driver.find_element_by_name("password")
    elem.send_keys(pwd)

    elem = driver.find_element_by_name("sub")
    elem.click()   

    elem = driver.find_element_by_name("job_list")

    html_tab_vol_brain=elem.get_attribute('innerHTML')
    txt_tab_vol_brain=elem.text
    driver.stop_client()       
    driver.close()
    deb=10
    for m in re.finditer(" "+str(ladate)+" ",txt_tab_vol_brain):
        deb -= 1
    return deb
```

```python
def getProceedTask(idvb, pwd, ladate, STUDY_PATH):
    #ladate=datetime.datetime.today().strftime('%Y-%m-%d')
    #print(ladate)
    #ladate="0520"
    options = Options()
    options.headless = True
    driver = Firefox(options=options)
    driver.implicitly_wait(15)
    #WebDriverWait(driver, timeout=5).until(document_initialised)
    driver.get("http://volbrain.upv.es/index.php")
    elem = driver.find_element_by_name("email")
    elem.send_keys(idvb)
    elem = driver.find_element_by_name("password")
    elem.send_keys(pwd)
    elem = driver.find_element_by_name("sub")
    elem.click()   
    elem = driver.find_element_by_name("job_list")
    html_tab_vol_brain=elem.get_attribute('innerHTML')
    txt_tab_vol_brain=elem.text

    for n in range(2,6,1):
        driver.execute_script("loadJobList("+str(n)+")")
        elem = driver.find_element_by_name("job_list")
        html_tab_vol_brain=html_tab_vol_brain+elem.get_attribute('innerHTML')
        txt_tab_vol_brain=txt_tab_vol_brain+elem.text

    #print(html_tab_vol_brain)
#https://files.volbrain.upv.es/b224399150720193121.zip
    regex="https:\/\/files.volbrain.upv.es.{1,100}?"+str(ladate)+".{1,10}?zip"
    #for m in re.finditer(" "+str(ladate)+" ",txt_tab_vol_brain):
    #    deb -= 1
    #print(html_tab_vol_brain)
    for it in re.finditer(regex,html_tab_vol_brain) :
        mystr=html_tab_vol_brain[it.start():it.end()]
        filepath=os.path.join(STUDY_PATH,'',mystr[30:])

        if not os.path.isfile(filepath) :
            attempts = 0
            while attempts < 3:
                try:
                    #response = urllib3.urlopen(mystr, timeout = 5)
                    #content = response.read()
                    
                    #http = urllib3.PoolManager(timeout=10.0)
                    #r = http.request('GET', mystr, retries=10)                   
                    
                    #f = open(filepath, 'w' )
                    #f.write( "coucou" )
                    #f.close() 
                    
                    r = requests.get(mystr, stream=True)
                    with open(filepath, 'wb') as f:
                        total_length = int(r.headers.get('content-length'))
                        for chunk in progress.bar(r.iter_content(chunk_size=1024), expected_size=(total_length/1024) + 1): 
                            if chunk:
                                f.write(chunk)
                                f.flush()                                      
                    
                    #!wget mystr
                    
                    break
                except urllib3.exceptions.NewConnectionError: #urllib3.URLError as e:
                    attempts += 1
                    print(type(e)) 
        
        if os.path.isfile(os.path.join(STUDY_PATH,'',mystr[30:44])): #(mystr[30:].startswith('1') | mystr[30:].startswith('0')) :
            #print(os.path.join(STUDY_PATH,'volbrain',mystr[30:44]))
            os.remove(os.path.join(STUDY_PATH,'',mystr[30:44])) 
        
    driver.stop_client()       
    driver.close()
```

```python
def sendFileOnVB(file,age,sexe,idvb,pwd):
    options = Options()
    options.headless = True
    driver = Firefox(options=options)
    driver.implicitly_wait(15)
   #WebDriverWait(driver, timeout=5).until(document_initialised)
    driver.get("https://volbrain.upv.es/index.php")
    #print("Web site reached")
    elem = driver.find_element_by_name("email")
    elem.send_keys(idvb)
    elem = driver.find_element_by_name("password")
    elem.send_keys(pwd)
    #print("Connected")
    elem = driver.find_element_by_name("sub")
    elem.click()
    elem = driver.find_element_by_id("pipeline1")
    elem.click()
    #print("Pipeline1 Ok")
    elem = driver.find_element_by_name("volbrain_t1_file")
    elem.send_keys(file)
    elem = driver.find_element_by_name("volbrain_patientssex")
    elem.send_keys(sexe)
    elem = driver.find_element_by_name("volbrain_patientsage")
    elem.send_keys(age)
    elem = driver.find_element_by_name("button_volbrain")
    elem.click()
    #print("Data sent")
    driver.stop_client()       
    driver.close()
```

# Envoyer les données

```python
img_fail = pd.DataFrame([], columns=('idvb', 'pwd', 'path', 'age', 'sexe', 'sub_id'))

for idvb, pwd in myDict.items():
    print(idvb)
    print(pwd)
    cmpt = 0
    avb = getNbAvailable(idvb,pwd)
    print(avb)
    for line in data.iterrows():
        if cmpt >= avb :
            break
        path = line[1]['t1_path']
        sexe = str(line[1]['sex'])
        age = int(line[1]['age'])
        sub_id = str(line[1]['sub_id']).replace('_', '')

        if not(np.isin(sub_id, already_proc)):
            cmpt += 1
            try :
                sendFileOnVB(path,age,sexe,idvb,pwd)
                print(sub_id)
            except :
                print('pas bon')
                img_fail = img_fail.append(pd.DataFrame([[idvb, 
                                                          pwd, 
                                                          path, 
                                                          age, 
                                                          sexe,
                                                          sub_id]], 
                                                       columns=('idvb', 
                                                                'pwd', 
                                                                'path',
                                                                'age',
                                                                'sexe',
                                                                'sub_id')))
                
                print(sub_id)
                print('fin pas bon')
            already_proc = np.append(already_proc, sub_id)
```

```python
path_down = '/NAS/deathrow/mri_data/Downloads/volbrain/'

for sub in glob(path_down+'/*'):
    for file in glob(sub+'/*'):
        try :
            if ('native' in file):
                if 'zip' in file :
                    if not(os.path.exists(sub+'/native')) :
                        os.system('mdkir '+sub+'/native')
                        zip_ref = zipfile.ZipFile(file, 'r')
                        zip_ref.extractall(sub+'/native')
                        os.remove(file)
            else : 
                if 'zip' in file :
                    if not(os.path.exists(sub+'/mni')) :
                        os.system('mkdir '+sub+'/mni')
                        zip_ref = zipfile.ZipFile(file, 'r')
                        zip_ref.extractall(sub+'/mni')
                        os.remove(file)
        except :
            print(sub)
```
