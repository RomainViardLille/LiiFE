```python
import nibabel as nib
import numpy as np
from scipy.ndimage import zoom
import matplotlib.pyplot as plt
```

```python
def z_score_normalize_image(image_path):
    
    image = nib.load(image_path)

    data = image.get_fdata()

    mean = np.mean(data)
    std = np.std(data)

    z_scores = (data - mean) / std

    # Create a new NIfTI image with the Z-score normalized data
    normalized_image = nib.Nifti1Image(z_scores, image.affine)

    return normalized_image
```

```python
def plot_histogram(image_path, bins=50):
    
    image = nib.load(image_path).get_fdata()

    min_value = np.min(image)
    max_value = np.max(image)

    plt.hist(image.ravel(), bins=bins, range=(min_value, max_value))
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.title('Histogram of MRI Image')
    plt.grid(True)
    plt.show()
```

```python
def plot_histogram2(image_path, bins=50):
    # Load the MRI image
    image = nib.load(image_path).get_fdata()
  
    filtered_values = image[image >= 0]
    
    plt.hist(filtered_values.ravel(), bins=bins, range=(0, np.max(filtered_values)))
    plt.title('Histogram of MRI Image (Values >= 0)')
    plt.grid(True)
    plt.show()
```

```python
def plot_combined_histogram(image1_path, image2_path, bins=50):
    # Load the MRI images
    image1 = nib.load(image1_path).get_fdata()
    image2 = nib.load(image2_path).get_fdata()

    filtered_values1 = image1[image1 >= 0]
    filtered_values2 = image2[image2 >= 0]

    min_value = min(np.min(filtered_values1), np.min(filtered_values2))
    max_value = max(np.max(filtered_values1), np.max(filtered_values2))

    plt.hist(filtered_values1.ravel(), bins=bins, range=(min_value, max_value), color='blue', alpha=0.5, label='Image 1')

    plt.hist(filtered_values2.ravel(), bins=bins, range=(min_value, max_value), color='red', alpha=0.5, label='Image 2')

    plt.title('Combined Histogram of Image 1 and Image 2 (Values >= 0)')
    plt.grid(True)
    
    plt.legend()

    plt.show()
```

```python
def plot_difference_histogram(image1_path, image2_path, bins=50):
    # Load the MRI images
    image1 = nib.load(image1_path).get_fdata()
    image2 = nib.load(image2_path).get_fdata()

    # Calculate the absolute difference between the two images
    absolute_difference = np.abs(image1 - image2)

    # Filter the data to exclude values below 0
    filtered_values = absolute_difference[absolute_difference >= 0]

    # Calculate the range for the histogram based on the min and max values in the filtered data
    min_value = np.min(filtered_values)
    max_value = np.max(filtered_values)

    # Plot a histogram of the absolute difference values in blue
    plt.hist(filtered_values.ravel(), bins=bins, range=(min_value, max_value), color='blue', alpha=0.7)
    plt.xlabel('Absolute Pixel Value Difference')
    plt.ylabel('Frequency')
    plt.title('Histogram of Absolute Difference between Image 1 and Image 2 (Values >= 0)')
    plt.grid(True)

    plt.show()
```

```python
def bland_altman_plot(data1, data2):
    
    image1 = nib.load(data1).get_fdata()
    image2 = nib.load(data2).get_fdata()
    
    mean = (image1 + image2) / 2
    diff = image1 - image2

    mean_diff = np.mean(diff)
    std_diff = np.std(diff, ddof=1)
    lower_limit = mean_diff - 1.96 * std_diff
    upper_limit = mean_diff + 1.96 * std_diff

    
    non_zero_diff = diff[diff != 0]

    #percent_non_zero_diff = len(non_zero_diff) / len(diff) * 100

    plt.figure(figsize=(8, 6))
    plt.scatter(mean, diff, c='b', alpha=0.5, s=30)
    plt.axhline(mean_diff, color='red', linestyle='--', linewidth=2)
    plt.axhline(lower_limit, color='green', linestyle='--', linewidth=1)
    plt.axhline(upper_limit, color='green', linestyle='--', linewidth=1)
    plt.xlabel('Moyenne des deux mesures')
    plt.ylabel('Diff√©rence entre les mesures')
    plt.title('Bland-Altman Plot')
    plt.grid(True)

    #plt.text(0.05, 0.1, f'Non-zero differences: {percent_non_zero_diff:.2f}%', transform=plt.gca().transAxes)
    
    # Show the plot
    plt.show()
```
