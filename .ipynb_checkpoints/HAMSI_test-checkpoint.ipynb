## Enlever les lignes inutiles 

```python
for x in ma_table.index : 
    if 'rfmt' in ma_table.loc[x,'SeriesDescription'] \
       or 'RFMT' in ma_table.loc[x,'SeriesDescription'] \
       or 'RFT' in ma_table.loc[x,'SeriesDescription']: 
        ma_table.drop(x, inplace = True)
```

```python
for x in ma_table.index :
    if 'DEFAULT PS SERIES' in ma_table.loc[x,'SeriesDescription'] \
       or 'no value' in ma_table.loc[x,'SeriesDescription'] \
       or 'Survey'in ma_table.loc[x,'SeriesDescription'] \
       or 'New' in ma_table.loc[x,'SeriesDescription'] \
       or 'Screen Save'in ma_table.loc[x,'SeriesDescription']:
        ma_table.drop(x,inplace = True)
```

## Liste des séries T1

```python
mypath = join('/NAS','dumbo','protocoles','HAMSI','data')
```

```python
table_T1 = ma_table[{'PatientID','SeriesDescription','StudyInstanceUID','SerieInstanceUID'}]
table_T1 = table_T1.reindex(columns = ['PatientID','SeriesDescription','StudyInstanceUID','SerieInstanceUID'])
```

```python
for x in table_T1.index :
    if 'T1' not in table_T1.loc[x,'SeriesDescription']:
        table_T1.drop(x,inplace = True)
```

```python
machines_T1 = []
Station_Name = []
Manufacturer_Model_Name = []

for x in table_T1.index : 
    serie = glob.glob(join(mypath,table_T1.loc[x,'StudyInstanceUID'],table_T1.loc[x,'SerieInstanceUID'],"*"),recursive = True)
    ds=dcm.dcmread(serie[0])
    
    machines_T1.append(ds.Manufacturer)
    Station_Name.append(ds.StationName)
    Manufacturer_Model_Name.append(ds.ManufacturerModelName)

table_T1['Machine'] = machines_T1
table_T1['Manufacturer Model Name'] = Manufacturer_Model_Name
table_T1['Station Name'] = Station_Name

nbT1_parmachine = table_T1.groupby('Station Name')['StudyInstanceUID'].nunique()
nbT1_parmachine
```

### Table des séries T1 sur GE

# Constructions de la table contenant les T1 avec Gado et faites sur la GE

```python
table_T1_Gado = [ma_table.loc[w] for w in ma_table.index if (('Sag CUBE T1 Gado' == ma_table.loc[w,'SeriesDescription'])or('Sag CUBE T1 gd' == ma_table.loc[w,'SeriesDescription']))]
df_table_T1_Gado = pd.DataFrame(table_T1_Gado)

Station_Name = []
for x in df_table_T1_Gado.index : 
    serie = glob.glob(join(mypath,df_table_T1_Gado.loc[x,'StudyInstanceUID'],df_table_T1_Gado.loc[x,'SerieInstanceUID'],"*"),recursive = True)
    ds=dcm.dcmread(serie[0])
    if 'DERIVED' in ds.ImageType :
        df_table_T1_Gado.drop(x,inplace = True)
    else :
        Station_Name.append(ds.StationName)


df_table_T1_Gado['Station Name'] = Station_Name    

table_T1_Gado_2 = df_table_T1_Gado[df_table_T1_Gado['Station Name'] == 'MR02']
df_table_T1_Gado_2=pd.DataFrame(table_T1_Gado_2)
len(df_table_T1_Gado_2)
```
