## DIFF Deferiprone Vs. Placebo à W36

```python
for col in ["original_firstorder_Median"]:
    #q_low = val[col].mean()-2*(val[col].std()) #val["GPe_L"].quantile(0.01)
    #q_hi  = val[col].mean()+2*(val[col].std()) #val["GPe_L"].quantile(0.99)
    #test = val_diff[(val_W00[col] < q_hi) & (val_W00[col] > q_low)]
    val_G1=val_R2_W36.query("GROUPE == 1")[col].values
    val_G2=val_R2_W36.query("GROUPE == 2")[col].values
    print("Deferiprone (mean/std)",val_G1.mean(),val_G1.std())
    print("Placebo (mean/std)",val_G2.mean(),val_G2.std())
    stat,pval=stats.mannwhitneyu(val_R2_W36.query("GROUPE == 1")[col].values,val_R2_W36.query("GROUPE == 2")[col].values)
    val_R2_W36['pvalue='] = np.round(pval,3).astype(str)
    print(np.round(pval,3).astype(str))
    plt.figure(figsize=(6,3))
    #print(stats.ttest_ind(test.query("GROUPE == 1")['SN_L'].values,test.query("GROUPE == 2")['SN_L'].values, equal_var=False))    
    sns.violinplot(data=val_R2_W36,y='pvalue=',x=col,split=True,hue="GROUPE",palette="pastel")
    sns.swarmplot(data=val_R2_W36,y='pvalue=',x=col,split=True,hue="GROUPE",palette="bright",alpha=0.8)
    FIG_FILE=os.path.join("/NAS","dumbo","protocoles","FAIRPARKII","W36_DefVsPlac_Median_SN.png")
    plt.savefig(FIG_FILE,dpi=300)
    plt.legend("")
    FIG_FILE=os.path.join("/NAS","dumbo","protocoles","FAIRPARKII","W36_DefVsPlac_Median_wolegend_SN.png")
    plt.savefig(FIG_FILE,dpi=300)
    print("col : ",col,val_R2_W36.shape[0])
```

# CALCUL DU DICE ENTRE LILLE ET INNSBRUCK

```python
from sklearn.metrics import jaccard_score
liste_filesLil=glob.glob(os.path.join("/NAS","dumbo","protocoles","FAIRPARKII",'Atlas_Seg','''*''','''SN_L*_on3DT1_std_ras_bin_r.nii.gz'''))
liste_filesLil.sort()
for file_lil in liste_filesLil:
    liste_filesInn=glob.glob(os.path.join("/NAS","dumbo","protocoles","FAIRPARKII",'export_romain_31-03-2022','labels','fc_dense_net_sigmoid',file_lil.split("/")[6]+"*.nii.gz"))
        #liste_files=glob.glob(os.path.join("/NAS","dumbo","protocoles","FAIRPARKII",'Atlas_Seg','''*''','''SN_*_on3DT1_std_ras_bin_r.nii.gz'''))
    
    lil1=nib.load(file_lil).get_fdata()
    ima_inn=nib.load(liste_filesInn[0]).get_fdata()
    
    lil2=nib.load(file_lil.replace("SN_L","SN_R")).get_fdata()
    
    lil3=lil1+lil2
    inn=np.logical_or(ima_inn==3,ima_inn==4)
    res3=jaccard_score(lil3.flatten(),inn.flatten())
    OVERLAP=np.logical_and(inn,lil3).sum()/np.min([inn.sum(),lil3.sum()])
    DSC=2*np.logical_and(inn,lil3).sum()/(inn.sum()+lil3.sum())
    JACC=np.logical_and(inn,lil3).sum()/np.logical_or(inn,lil3).sum()
    print(file_lil.split('/')[6],OVERLAP,DSC,JACC)
```
