```python
def unconfound(y, confound, group_data=False):
    """
    This will remove the influence "confound" has on "y".
    If the data is made up of two groups, the group label (indicating the group) must be the first column of
    'confound'. The group label will be considered when fitting the linear model, but will not be considered when
    calculating the residuals.
    Args:
        y: [samples, targets]
        confound: [samples, confounds]
        group_data: if the data is made up of two groups (e.g. for t-test) or is just
                    one group (e.g. for correlation analysis)
    Returns:
        y_correct: [samples, targets]
    """
    # Demeaning beforehand or using intercept=True has similar effect
    #y = demean(y)
    #confound = demean(confound)

    lr = LinearRegression(fit_intercept=True).fit(confound, y)  # lr.coef_: [targets, confounds]
    if group_data:
        y_predicted_by_confound = lr.coef_[:, 1:] @ confound[:, 1:].T
    else:
        y_predicted_by_confound = lr.coef_ @ confound.T  # [targets, samples]
    y_corrected = y.T - y_predicted_by_confound
    return y_corrected.T  # [samples, targets]
```

```python
class FeaturesCorrDel:
    def __init__(self, threshold=0.95):
        self.threshold = threshold
    
    def fit(self,X,Y):
        # ?liminaton pr?alable des features constantes
        self.mask_ = np.std(X, axis=0) != 0
        
        # collecte paires de features corr?l?es
        corr = []
        corr_matrix = abs(np.cov(X,rowvar=False))
        for i in range(X.shape[1]-1):
            if not self.mask_[i]: continue
            for j in range(i+1, X.shape[1]):
                if self.mask_[j] and corr_matrix[i][j] > self.threshold:
                    corr.append((i,j,corr_matrix[i][j]))
        
        # tri des paires de features corr?l?es
        corr = sorted(corr, key=lambda entry: -entry[2])
        
        # suppession features
        while corr:
            entry = corr[0]
            corr_1y = abs(np.mean([X[:,entry[0]][k] * Y[k] for k in range(Y.shape[0])]))
            corr_2y = abs(np.mean([X[:,entry[1]][k] * Y[k] for k in range(Y.shape[0])]))
            to_del = entry[0] if corr_1y < corr_2y else entry[1]
            self.mask_[to_del] = False
            corr = [c for c in corr if to_del not in (c[0],c[1])]
        return self

    def transform(self,X):
        return X[:,self.mask_]

    def fit_transform(self,X,Y):
        self.fit(X,Y)
        return self.transform(X)
```

### MultiTaskElasticNetCV

```python
mycolumns=[]
selectedFeatures=[]
for col in Cli_Gen_data.columns:
    if ((col.find('3DT1')>-1)) and ((col.find("_SubstanciaNigra")>-1) or (col.find("_Caudate")>-1) or (col.find("_Putamen")>-1) or (col.find("_Striatum")>-1) or (col.find("_GPi")>-1)):
    #if (col.find('3DT1')>-1) and ((col.find('Percentile')>-1) or (col.find('Entropy')>-1)) and ((col.find("_SubstanciaNigra")>-1) or (col.find("_Caudate")>-1) or (col.find("_Putamen")>-1) or (col.find("_Striatum")>-1) or (col.find("_GPi")>-1)):
        mycolumns.append(col)
        
liste_cov=['DOPA']#Cli_Gen_data.columns[7:16]#,'SNP_ACSL4'
    
mysetcol=set(mycolumns)
for tar in liste_cov:
    mysetcol.add(tar)
mysetcol.add('Age')
mysetcol.add('Duree')

X = Cli_Gen_data.dropna(axis=0,how='any',subset=mysetcol,inplace=False)[mycolumns].values

# Create y from output
y = Cli_Gen_data.dropna(axis=0,how='any',subset=mysetcol,inplace=False)[liste_cov].values
print(y.shape)
EV=[Cli_Gen_data.dropna(axis=0,how='any',subset=mysetcol,inplace=False)['Age'].values,
Cli_Gen_data.dropna(axis=0,how='any',subset=mysetcol,inplace=False)['Duree'].values]

X_norm=StandardScaler().fit(X).transform(X)

sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
X_sel1=sel.fit_transform(X_norm)

X_corr=unconfound(X_sel1,np.transpose(np.asarray(EV)),False)

#print(X.shape,X_sel1.shape)

clf = linear_model.MultiTaskElasticNetCV(cv=3,max_iter=15000,normalize=True,verbose=1,n_jobs=8)
clf.fit(X_corr,y)
```

```python
from sklearn.metrics import r2_score
print(clf.coef_.shape)
print(np.max(clf.coef_,axis=0),np.argmax(clf.coef_,axis=0))
r2_score(y,clf.predict(X_corr))
```
