```python
# Importation des différentes bibliothèques
import sys
import os
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Récuperation des chemins d'accès des différents fichiers
file1=os.path.join("/NAS","dumbo","protocoles","FAIRPARKII","AP","QSMdataAnxAP.csv")
file2=os.path.join("/NAS","dumbo","protocoles","FAIRPARKII","AP","R2dataAnxAP.csv")
file3=os.path.join("/NAS","dumbo","protocoles","FAIRPARKII","AP","DataAnxAP.xlsx")
file4=os.path.join("/NAS","dumbo","protocoles","FAIRPARKII","AP","PtsWithCQR2Ok.csv")

# Affichage des chemins d'accès des différents fichiers recuperés
print(file1)
print(file2)
print(file3)
print(file4)

# Importation des différents fichiers sous forme de dataframe à partir des chemins d'accès recuperés plus haut
val1=pd.read_csv(file1)
val2=pd.read_csv(file2)
val3=pd.read_excel(file3, engine='openpyxl')
val4=pd.read_csv(file4)

# Utilisation de la fonction merge pour recuperer uniquement les informations des patients présents dans le fichier nommé DataAnxAp.xlsx (intersection) 
result1=val1.merge(val2,on="id_patient")
result2=val3.merge(result1,on="id_patient")
result3=val1.merge(val4, how='inner')

idlist2=list(result2['id_patient']) # Recuperation sous forme de liste des identifiants des patients possédant à la fois des valeurs de R2, QSM et présents dans le fichier DataAnxAp.xlsx
idlist3=list(result3['id_patient']) # Recuperation sous forme de liste des identifiants des patients presents dans le fichier DataAnxAp.xlsx 
         
for idpat in idlist2:
    if idpat in idlist3:          # On teste si les identifiants des patients presents dans l'idlist2 sont egalement presenrts dans l'idlist3
        ind=idlist2.index(idpat)  # Recuperation de l'index ou position de l'identifiant dans la liste idlist2
        result2.loc[ind,'CQ']='1'  # Affectation de la valeur 1 à la colonne CQ ( Controle qualité ok)
    else:
        ind1=idlist2.index(idpat)
        result2.loc[ind1,'CQ']='0'    # Affectation de la valeur 0 à la colonne CQ (Controle qualité pas ok)
result2       
```

```python
gfg_csv_data=result2.to_csv('parkinson.csv', index = True) # Conversion et création d'un fichier csv ayant pour nom "parkinson.csv"
df2=result2.select_dtypes(include=['int64','float64']) #Filtre le dataframe de facon à extraire uniquement les colonnes de type entier et réel
scorez=stats.zscore(df2, nan_policy='omit')  # calcul des zscores
scorez                                       # Affichage de la dataframe uniquement avec les zscores calculés
```

```python
result2_nona=result2.dropna()     # supprimer toutes les lignes contenant la mention "nan" du dataframe
result2_nona                      # Affichage de la dataframe sans les lignes contenant la mention "nan"
```

```python
nomcolumn=result2_nona.columns.tolist()
del(nomcolumn[0 : 6]) # On supprime les éléments des colonnes d’indices 0 à 6 (6 exclu).
print(nomcolumn)
for mastruct in nomcolumn:
    y0=list(result2_nona[mastruct])
    x0=list(result2_nona['MDS 1.4'])
    x=np.array(x0)
    y=np.array(y0)
    (a, b, r, val1, val2)= stats.linregress(x, y)  # retourne les paramètres de la droite de regression
    A=round(a,4)
    B=round(b,4)
    R=round(r**2,4)
    print(mastruct,A,B,R)                    # a=coéfficient directeur  b=ordonnée à l'origine  r**2= coéfficient de regression linéaire 
    plt.plot(x, y, "b.", label="donnees") # les coordonnées (x, y) representés par des points
    y_predict=a*x+b
    plt.plot(x, y_predict, "-r",label="regression")  # droite de regression
    plt.xlabel("MDS 1.4") # nom de l'axe des abcisses
    plt.ylabel(mastruct) # nom de l'axe des ordonnées 
    plt.legend() # la legende
    plt.title(mastruct) # titre de graphique
    plt.show()  # afficher les courbes
    
```
