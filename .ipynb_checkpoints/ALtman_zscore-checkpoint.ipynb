```python
import nibabel as nib
import numpy as np
from scipy.ndimage import zoom
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
def bland_altman_plot(data1, data2):
    # Calculate the mean and the difference between the two images
    mean = (data1 + data2) / 2
    diff = data1 - data2

    # Calculate the mean difference and the limits of agreement
    mean_diff = np.mean(diff)
    std_diff = np.std(diff, ddof=1)
    lower_limit = mean_diff - 1.96 * std_diff
    upper_limit = mean_diff + 1.96 * std_diff

    # Exclude zero differences
    #non_zero_diff = diff[diff != 0]

    # Calculate the percentage of non-zero differences
    #percent_non_zero_diff = len(non_zero_diff) / len(diff) * 100

    # Create the Bland-Altman plot
    plt.figure(figsize=(8, 6))
    plt.scatter(mean, diff, c='b', alpha=0.5, s=30)
    plt.axhline(mean_diff, color='red', linestyle='--', linewidth=2)
    plt.axhline(lower_limit, color='green', linestyle='--', linewidth=1)
    plt.axhline(upper_limit, color='green', linestyle='--', linewidth=1)
    plt.xlabel('Moyenne des deux mesures')
    plt.ylabel('DiffÃ©rence entre les mesures')
    plt.title('Bland-Altman Plot')
    plt.grid(True)

    # Display the percentage of non-zero differences
    #plt.text(0.05, 0.1, f'Non-zero differences: {percent_non_zero_diff:.2f}%', transform=plt.gca().transAxes)
    
    # Show the plot
    plt.show()
```

```python
def plot_normalized_image_histogram(normalized_image):
    # Get voxel values as a numpy array from the normalized image
    data = normalized_image.get_fdata()

    # Flatten the data array to a 1D array
    flattened_data = data.ravel()

    # Create a histogram of the flattened data
    #plt.hist(flattened_data, bins=256, range=(0, 500), color='blue', alpha=0.7)
    plt.hist(flattened_data,50,range=(10,500), color='blue', alpha=0.7)

    # Customize the histogram plot
    plt.title('Histogram of Z-Score Normalized Image')
    plt.xlabel('Pixel Value (Z-Score)')
    plt.ylabel('Frequency')

    # Show the histogram plot
    plt.show()
```

```python
image_file1 = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Brains/CHU/Pat8_T1Gado_CHU_brain.nii.gz"
image_file2 = "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Brains_Registered/COL_registred_on_CHU/Pat8_COL_registered.nii.gz"

# Load NIfTI images
image1 = nib.load(image_file1).get_fdata()
image2 = nib.load(image_file2).get_fdata()

bland_altman_plot(image1 , image2)
```

```python
def z_score_normalize_image(image_path):
    
    image = nib.load(image_path)

    data = image.get_fdata()

    mean = np.mean(data>0)
    std = np.std(data>0)

    z_scores = (data - mean) / std

    # Create a new NIfTI image with the Z-score normalized data
    normalized_image = nib.Nifti1Image(z_scores, image.affine)

    return normalized_image
```

```python
normalized_image = z_score_normalize_image("/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Brains/CHU/Pat8_T1Gado_CHU_brain.nii.gz")
nib.save(normalized_image, "/NAS/dumbo/protocoles/HAMSI/OurData4Mispel_T1Gado/Trash/normalized_image.nii.gz")
```
